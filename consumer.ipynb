```python
from kafka import KafkaConsumer
import json

consumer = KafkaConsumer(
    'ztm_pozycje_gps',
    bootstrap_servers='broker:29092',
    value_deserializer=lambda v: json.loads(v.decode('utf-8')),
    key_deserializer=lambda k: k.decode('utf-8') if k else None,
    auto_offset_reset='latest',
    consumer_timeout_ms=30000,
    group_id='podglad-debug'
)

print("Czekam na wiadomości z topicu 'ztm_pozycje_gps'...")
print("(jak producer działa, powinny pojawić się w ciągu 15s)\n")

count = 0
for message in consumer:
    print(f"=== Wiadomość #{count + 1} ===")
    print(f"Klucz (VehicleNumber): {message.key}")
    print(f"Partition: {message.partition}, Offset: {message.offset}")
    print(f"Wartość:")
    print(json.dumps(message.value, indent=2, ensure_ascii=False))
    print()
    
    count += 1
    if count >= 5:
        break

consumer.close()
print(f"Pokazano {count} wiadomości.")
```
