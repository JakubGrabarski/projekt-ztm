```python
import psycopg2

try:
    conn = psycopg2.connect(
        host="postgres",
        port=5432,
        database="ztm_data",
        user="postgres",
        password="ztm123"
    )
    cursor = conn.cursor()
    
    # 1. Liczba rekordów w pojazdy_aktualne
    cursor.execute("SELECT COUNT(*) FROM pojazdy_aktualne;")
    count_aktualne = cursor.fetchone()[0]
    
    # 2. Liczba rekordów w pozycje_historia
    cursor.execute("SELECT COUNT(*) FROM pozycje_historia;")
    count_historia = cursor.fetchone()[0]
    
    print("==================================================")
    print(f"📊 STAN BAZY DANYCH ZTM:")
    print("==================================================")
    print(f"🚕 Unikalne pojazdy (pojazdy_aktualne): {count_aktualne}")
    print(f"📈 Wszystkie punkty GPS (pozycje_historia): {count_historia}")
    print("==================================================")
    
    cursor.close()
    conn.close()

except Exception as e:
    print(f"Błąd podczas liczenia rekordów: {e}")
```
